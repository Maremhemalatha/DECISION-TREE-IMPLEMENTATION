```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier,plot_tree 
from sklearn.preprocessing import LabelEncoder 
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix 
import seaborn as sns
from IPython.display import display,HTML
df=pd.read_csv("C:\\Users\\marem\\Downloads\\zoo.csv")
display(HTML("<mark>ZOO DATASET</mark>"))
print(df.head())
display(HTML("<mark>ZOO DATASET INFORMATION</mark>"))
print(df.info())

x=df.drop(['animal_name','class_type'],axis=1)
y=df['class_type']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=DecisionTreeClassifier(random_state=42)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
display(HTML("<mark>Accuracy</mark>"))
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Classification Report:\n",classification_report(y_test,y_pred,zero_division=0))
plt.figure(figsize=(20,10))
plot_tree(model,filled=True,feature_names=x.columns,class_names=[str(i) for i in sorted(df['class_type'].unique())])
plt.title("Zoo Animal Classification Decision Tree")
plt.show()
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
##Confusion Matrix Analysis
##The Decision Tree Classifier achieved **perfect Classification** on the test data.
##All instances were correctly classified,as seen from the diagonal-only confusion matrix.
##This indicates that the selected features are strong and the decision boundaries sre well-learned by the tree
```
